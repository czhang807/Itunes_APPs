## 5. Analyze the data

**5.i) Clean the data.**

```python
import pandas as pd
data = pd.read_csv("3.csv")
data.head()
```

```python
# Some rows do not pull in star_rating and num_ratings 
# Drop these rows for the analysis
data = data.dropna()
data.describe(include='all')
```

In this part, I imported the csv file and delected null values. I also looked at the decriptive info for each column

**5.ii**

```python
data.info()
```

```python
# Create a clean ratings column converts 189.5K to 189500 etc
number_ratings_cleaned = data['number_rating'].apply(lambda x: x[:-8].strip())
data['num_ratings_clean'] = [float(i[:-1])*1000000 if i[-1] == 'M' else float(i[:-1])*1000 if i[-1] == 'K'else float(i) for i in list(number_ratings_cleaned)]
data['num_ratings_clean'] = data['num_ratings_clean'].astype(int)
```

```python
data
```

Changed the format and datatype of the "number of ratings" column

**5.iii**

```python
data.sort_values(by = ['star_rating', 'num_ratings_clean', 'app_name'], ascending=[False, False, True])\
[['app_name','star_rating','num_ratings_clean']].head(10)
```

Listed the names of the top 10 apps in descending order based on star rating and within those with the same
star rating sort based on number of ratings in descending order. If the number of ratings are also the same, sorted
by app_name in ascending order.

**5.iv**

```python
data.groupby(by = 'category', as_index = 'category').size().sort_values(ascending = False)
```

Listed the number of apps for each category

**5.v**

```python
data.groupby(by = 'category', as_index = 'category').mean()['star_rating'].sort_values(ascending = False)
```

Listed the average rating of all apps in that category and sorted in descending order by average rating.
